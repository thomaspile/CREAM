```python
macd_fastp = 12
macd_slowp = 26
macd_sigp = 9

class MACD:
    
    def __init__(self, series, fastp=12, slowp=26, sigp=9):
        
        self.series = series

        self.macd, self.macd_signal, self.macdHist = talib.MACD(self.series, fastperiod=fastp, 
                                                                slowperiod=slowp, signalperiod=sigp)
    
    def scale_to_series(self):
        
        '''Only used if you want to plot the MACD line and signal on the series plot'''
    
        scaler = MinMaxScaler(feature_range=(min(self.series), max(self.series)))
        scaler.fit(self.macd.reshape(-1, 1))

        self.macd_scaled = scaler.transform(self.macd.reshape(-1, 1))
        self.macd_signal_scaled = scaler.transform(self.macd_signal.reshape(-1, 1))
    
    def get_crossovers(self):
        
        self.crossovers = ['None']
        self.idxs_up = []
        self.idxs_down = []

        for i in range(1, len(self.macd_signal)):
            if self.macd[i] > self.macd_signal[i] and self.macd[i - 1] <= self.macd_signal[i - 1]:
                self.crossovers.append("Up")
                self.idxs_up.append(i)
            elif self.macd[i] < self.macd_signal[i] and self.macd[i - 1] >= self.macd_signal[i - 1]:
                self.crossovers.append("Down")
                self.idxs_down.append(i)
            else:
                self.crossovers.append('None')
                
    def create_dataframe(self):
        
        df_up_signals = pd.DataFrame({'crossover_idx':self.idxs_up, 'crossover_type':['up']*len(self.idxs_up)})
        df_down_signals = pd.DataFrame({'crossover_idx':self.idxs_down, 'crossover_type':['down']*len(self.idxs_down)})
        
        self.df_macd = pd.concat([df_up_signals, df_down_signals]).sort_values('crossover_idx', ascending=True)


class Peaks:
    
    # Define type of peak (possibly with prominence, but depends whether scipy uses much forward view for this)
    # Assess changes between peaks (tops and bottoms)
    
    def __init__(self, prices):
        
        self.prices = prices
        self.df_prices = pd.DataFrame({'idx':range(0, len(self.prices)), 'price':self.prices})
        
    def get(self, prominence=2):
        
        top_idxs = find_peaks(self.prices, prominence=prominence)
        bottom_idxs = find_peaks(-self.prices, prominence=prominence)
        top_idxs = list(top_idxs[0])
        bottom_idxs = list(bottom_idxs[0])        
    
    def create_relationships(self, n_peaks_back=3):
        
        df_tops = pd.DataFrame({'idx':top_idxs, 'type':['top']*len(top_idxs)})
        df_bottoms = pd.DataFrame({'idx':bottom_idxs, 'type':['bottom']*len(bottom_idxs)})
        
        self.df_peaks = pd.concat([df_tops, df_bottoms]).sort_values('idx', ascending=True)
        self.df_peaks = self.df_peaks.merge(self.df_prices, on='idx', how='left')
        self.df_peaks = self.df_peaks.rename(columns={'price':'peak_price', 'idx':'peak_idx', 'type':'peak_type'})
        
        for i in range(1, 2 * n_peaks_back + 1):
            self.df_peaks[f'peak_idx_minus_{i}'] = self.df_peaks['peak_idx'].shift(i)
            self.df_peaks[f'peak_price_minus_{i}'] = self.df_peaks['peak_price'].shift(i)
            self.df_peaks[f'peak_type_minus_{i}'] = self.df_peaks['peak_type'].shift(i)
        
        for i in range(2, 2 * n_peaks_back + 2, 2):
            if i == 2:
                self.df_peaks[f'peak_same_type_change_prev_{int(i/2)}'] = self.df_peaks['peak_price'] - self.df_peaks[f'peak_price_minus_{i}']
            else:            
                self.df_peaks[f'peak_same_type_change_prev_{int(i/2)}'] = self.df_peaks[f'peak_price_minus_{i-2}'] - self.df_peaks[f'peak_price_minus_{i}'] 
  
    def find_trending_peaks(self, min_proportion=0.05):
        
        self.df_peaks['ascending'] = np.where(self.df_peaks[f'peak_same_type_change_prev_1'] > 0, 1, 0)
        self.df_peaks['ascending_double'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] > 0), 
                                                     1, 0)
        self.df_peaks['ascending_triple'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_3'] > 0), 
                                                     1, 0)

        self.df_peaks['descending'] = np.where(self.df_peaks[f'peak_same_type_change_prev_1'] < 0, 1, 0)
        self.df_peaks['descending_double'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] < 0), 
                                                     1, 0)
        self.df_peaks['descending_triple'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_3'] < 0), 
                                                     1, 0)
        
        
        # self.df_peaks['ascending_channel'] = np.where()
```
