```python
macd_fastp = 12
macd_slowp = 26
macd_sigp = 9

class MACD:
    
    def __init__(self, series, periods=None, fastp=12, slowp=26, sigp=9):
        
        self.series = series
        self.periods = periods

        self.macd, self.macd_signal, self.macdHist = talib.MACD(self.series, fastperiod=fastp, 
                                                                slowperiod=slowp, signalperiod=sigp)
    
    def scale_to_series(self):
        
        '''Only used if you want to plot the MACD line and signal on the series plot'''
    
        scaler = MinMaxScaler(feature_range=(min(self.series), max(self.series)))
        scaler.fit(self.macd.reshape(-1, 1))

        self.macd_scaled = scaler.transform(self.macd.reshape(-1, 1))
        self.macd_signal_scaled = scaler.transform(self.macd_signal.reshape(-1, 1))
    
    def get_crossovers(self):
        
        self.crossovers = ['None']
        self.idxs_up = []
        self.idxs_down = []

        for i in range(1, len(self.macd_signal)):
            if self.macd[i] > self.macd_signal[i] and self.macd[i - 1] <= self.macd_signal[i - 1]:
                self.crossovers.append("Up")
                self.idxs_up.append(i)
            elif self.macd[i] < self.macd_signal[i] and self.macd[i - 1] >= self.macd_signal[i - 1]:
                self.crossovers.append("Down")
                self.idxs_down.append(i)
            else:
                self.crossovers.append('None')
                
        df_up_signals = pd.DataFrame({'crossover_idx':self.idxs_up, 'crossover_type':['up']*len(self.idxs_up)})
        df_down_signals = pd.DataFrame({'crossover_idx':self.idxs_down, 'crossover_type':['down']*len(self.idxs_down)})
        
        self.df_macd_crossovers = pd.concat([df_up_signals, df_down_signals]).sort_values('crossover_idx', ascending=True)
 
    def get_divergence(self):
        
        self.df_macd = pd.DataFrame({'macd':self.macd, 'macd_signal':self.macd_signal})
        
        if self.periods is not None:
            self.df_macd['idx'] = self.periods
        else:
            self.df_macd = self.df_macd.reset_index().rename(columns={'index':'idx'})         
        
        self.df_macd['macd_signal_divergence'] = self.df_macd['macd'] - self.df_macd['macd_signal']        
    
    def get_divergence_features(self, n_back_checks=10, n_periods_gradient=4):
        
        for i in range(1, n_back_checks + 1):
            self.df_macd[f'macd_signal_divergence_minus_{i}'] = self.df_macd['macd_signal_divergence'].shift(i) 
        
        self.df_macd[f'macd_signal_divergence_gradient_{n_periods_gradient}'] = (self.df_macd[f'macd_signal_divergence'] - self.df_macd[f'macd_signal_divergence_minus_{n_periods_gradient}']) / n_periods_gradient
        self.df_macd[f'macd_signal_divergence_rolling_std_{n_back_checks}_delta'] = (self.df_macd['macd_signal_divergence'] / self.df_macd['macd_signal_divergence'].rolling(window=n_back_checks, closed='left').std()) - 1
        
        # self.df_macd[self.df_macd['idx'].isin(self.df_macd_crossovers['idx'].values
            
            
class Peaks:
    
    def __init__(self, prices):
        
        self.prices = prices
        self.df_prices = pd.DataFrame({'idx':range(0, len(self.prices)), 'price':self.prices})
        
    def get_all(self, prominence=None, threshold=None):
        
        top_idxs = find_peaks(self.prices, prominence=prominence, threshold=threshold)
        bottom_idxs = find_peaks(-self.prices, prominence=prominence, threshold=threshold)
        self.top_idxs = list(top_idxs[0])
        self.bottom_idxs = list(bottom_idxs[0])    
        
    def get_rolling(self, prominence=2):
        
        """ 
        This rolling process ensures that when defining peaks we are not using any future data from the current point in time
        """
        
        self.df_rolling_peaks = pd.DataFrame([self.rolling_latest_peak(df_rolling, prominence=prominence) for df_rolling in self.df_prices.rolling(self.df_prices.shape[0])])
        self.df_rolling_peaks.columns = ['peak_idx_latest']
        self.df_rolling_peaks = pd.concat([self.df_prices[['idx', 'price']], self.df_rolling_peaks], axis=1)
           
    def rolling_latest_peak(self, df_rolling, prominence=3, threshold=None):
        """
        When iterating over a rolling window it disregards the min_periods
        argument of rolling and will produce DataFrames for all windows

        The input is also of type DataFrame not Series

        You are completely responsible for doing all operations here,
        including ignoring values if the input is not of the correct shape
        or format

        """

        top_idxs = find_peaks(df_rolling['price'].values, prominence=prominence, threshold=threshold)
        bottom_idxs = find_peaks(-df_rolling['price'].values, prominence=prominence, threshold=threshold)
        top_idxs = list(top_idxs[0])
        bottom_idxs = list(bottom_idxs[0])        
        top_idxs.reverse()   
        bottom_idxs.reverse()
        
        all_peak_idxs = sorted(top_idxs + bottom_idxs)
        all_peak_idxs.reverse()
        if len(all_peak_idxs) > 0: latest_peak_idx = all_peak_idxs[0]
        else: latest_peak_idx = np.nan

        return latest_peak_idx
    
    def create_relationships(self, n_peaks_back=3):
        
        df_tops = pd.DataFrame({'idx':self.top_idxs, 'type':['top']*len(self.top_idxs)})
        df_bottoms = pd.DataFrame({'idx':self.bottom_idxs, 'type':['bottom']*len(self.bottom_idxs)})
        
        self.df_peaks = pd.concat([df_tops, df_bottoms]).sort_values('idx', ascending=True)
        self.df_peaks = self.df_peaks.merge(self.df_prices, on='idx', how='left')
        self.df_peaks = self.df_peaks.rename(columns={'price':'peak_price', 'idx':'peak_idx', 'type':'peak_type'})
        
        for i in range(1, 2 * n_peaks_back + 1):
            self.df_peaks[f'peak_idx_minus_{i}'] = self.df_peaks['peak_idx'].shift(i)
            self.df_peaks[f'peak_price_minus_{i}'] = self.df_peaks['peak_price'].shift(i)
            self.df_peaks[f'peak_type_minus_{i}'] = self.df_peaks['peak_type'].shift(i)
        
        for i in range(2, 2 * n_peaks_back + 2, 2):
            if i == 2:
                self.df_peaks[f'peak_same_type_change_prev_{int(i/2)}'] = self.df_peaks['peak_price'] - self.df_peaks[f'peak_price_minus_{i}']
            else:            
                self.df_peaks[f'peak_same_type_change_prev_{int(i/2)}'] = self.df_peaks[f'peak_price_minus_{i-2}'] - self.df_peaks[f'peak_price_minus_{i}'] 
  
    def find_trending_peaks(self, min_proportion=0.05):
        
        self.df_peaks['ascending'] = np.where(self.df_peaks[f'peak_same_type_change_prev_1'] > 0, 1, 0)
        self.df_peaks['ascending_double'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] > 0), 
                                                     1, 0)
        self.df_peaks['ascending_triple'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] > 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_3'] > 0), 
                                                     1, 0)

        self.df_peaks['descending'] = np.where(self.df_peaks[f'peak_same_type_change_prev_1'] < 0, 1, 0)
        self.df_peaks['descending_double'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] < 0), 
                                                     1, 0)
        self.df_peaks['descending_triple'] = np.where((self.df_peaks[f'peak_same_type_change_prev_1'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_2'] < 0)\
                                                     & (self.df_peaks[f'peak_same_type_change_prev_3'] < 0), 
                                                     1, 0)
        
    def merge_peak_stats(self):
        
        
        self.df_rolling_peaks = self.df_rolling_peaks.merge(self.df_peaks.rename(columns={'peak_idx':'peak_idx_latest'}), 
                                                            on='peak_idx_latest', how='left')
        self.df_rolling_peaks = self.df_rolling_peaks.rename(columns={ 
                                          'peak_type':'peak_type_latest', 
                                          'peak_price':'peak_price_latest',
                                          'ascending':'peak_ascending_latest',
                                          'ascending_double':'peak_ascending_double_latest',
                                          'ascending_triple':'peak_ascending_triple_latest',
                                          'descending':'peak_descending_latest',
                                          'descending_double':'peak_descending_double_latest',
                                          'descending_triple':'peak_descending_triple_latest',
                                         })
        
        self.df_rolling_peaks['change_from_latest_peak'] = self.df_rolling_peaks['price'] - self.df_rolling_peaks['peak_price_latest']
        self.df_rolling_peaks['periods_since_latest_peak'] = self.df_rolling_peaks['idx'] - self.df_rolling_peaks['peak_idx_latest']
        
        
        # self.df_peaks['ascending_channel'] = np.where()

        
```

```python
space_lgb = {
    'n_estimators': pyll.scope.int(hp.quniform('n_estimators', 10, 150, 10)),
    'boosting_type': hp.choice('boosting_type', ['gbdt', 'dart', 'goss']),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.5)),
    'max_depth': pyll.scope.int(hp.quniform('max_depth', 3, 16, 1)),
    # 'num_leaves': pyll.scope.int(hp.quniform('num_leaves', 2, 150, 1)),
    'min_child_samples': pyll.scope.int(hp.choice('min_child_samples', [5, 10, 20, 30, 40, 50, 60, 70, 80])),
    'reg_alpha': hp.choice('reg_alpha', [0, 0.001, 0.01, 0.1, 0.2]),
    'reg_lambda': hp.choice('reg_lambda', [0, 0.001, 0.01, 0.1, 0.2]),
    'cutoff': hp.quniform('cutoff', 0.01, 1, 0.01),
    # 'outcome_var': hp.choice('outcome_var', vars)
#     'feature_fraction': hp.uniform('feature_fraction', 0.1, 1.0),
#     'subsample': hp.uniform('subsample', 0.1, 1.0),
#     'subsample_for_bin': hp.quniform('subsample_for_bin', 20000, 300000, 20000),
    #,'colsample_bytree': hp.uniform('colsample_bytree', 0.6, 1.0)
}

space_logistic_regularised = {
  'penalty':hp.choice('penalty', ['l1']),
  'solver':hp.choice('solver', ['saga', 'liblinear']),  
  'C': hp.quniform('C', 0.01, 1, 0.01),
  'cutoff': hp.quniform('cutoff', 0.01, 1, 0.01)
}


def calculate_profit(estimator, X, y, cutoff=0.1):
    
    y_hat = estimator.predict_proba(X)[:, 1] 
    
    df = pd.DataFrame({'actual': y, 'pred': y_hat})
    df['buy'] = np.where(df['pred'] >= cutoff, 1, 0)
    
    df_buys = df[df.buy==1].copy()
    
    # buy_correct_rate = df_buys['actual'].mean()
    
    df_buys['outcome'] = np.where(df_buys['actual'] == 1, 1, -1)
    
    net_successful = df_buys['outcome'].sum()
    
    return - net_successful

def calculate_precision(estimator, X, y, cutoff=0.1):
    
    y_hat = estimator.predict_proba(X)[:, 1] 
    
    df = pd.DataFrame({'actual': y, 'pred': y_hat})
    df['buy'] = np.where(df['pred'] >= cutoff, 1, 0)
    
    df_buys = df[df.buy==1].copy()
    
    precision = df_buys['actual'].mean()
    
    return - precision


opt = OptimalModel(df_train=df_train,
                   df_test=df_test,
                   cols=feat_vars, 
                   outcome_var='next_price_up',
                   model_type='logistic', 
                   evals=20, 
                   opt_lib='hp', 
                   search_space=space_logistic_regularised, 
                   how_to_tune='cross_val_standardised', 
                   eval_metric=calculate_precision,
                   n_jobs=-1, 
                   seed=123, 
                   outcome_type='classification', 
                   k=5,
                   stratify_kfold=False)

model, params, trials, importance, errors, cv_scores = opt.build_optimal_model()
importance
```
